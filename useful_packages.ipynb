Zapoznajemy się z: numpy, matplotlib, plotly, seaborn, pandas

```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
def ciekawa_funkcja(x, y):
    return np.sin(x + y)  # jest też wersja w bibliotece math
```

```python
input_X = np.arange(-5.0, 5.0, 0.01)
input_Y = np.arange(-5.0, 5.0, 0.01)
X, Y = np.meshgrid(input_X, input_Y)
Z = ciekawa_funkcja(X, Y)
```

Matplotlib:

```python
fig = plt.figure(figsize=(18, 6))

"""Wykres 3D"""
ax1 = fig.add_subplot(121, projection='3d')  # 1x2 grid, pos. 1
ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_title('Wykres sin(x+y) 3D z matplotlib')
ax1.set_xlabel('Oś 0x')
ax1.set_ylabel('Oś 0y')
ax1.set_zlabel('Wartości funkcji')

"""Wykres 2D - rzut na płaszczyznę XY"""
ax2 = fig.add_subplot(122)  # 1x2 grid, pos. 2
contour_xy = ax2.contourf(input_X, input_Y, Z, cmap='viridis')
fig.colorbar(contour_xy, ax=ax2)
ax1.set_title('Rzut wartości funkcji sin(x+y) na płaszczyznę XY z matplotlib')
ax1.set_xlabel('Oś 0x')
ax1.set_ylabel('Oś 0y')

plt.tight_layout()
plt.show()
```
